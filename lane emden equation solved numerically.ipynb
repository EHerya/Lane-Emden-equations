```python
%pylab inline
import seaborn as sns
sns.set_palette("Set1",8)
```

```python
dxi=0.001
N=10000 #bins

def f(N):
    L1_5=[]
    L3=[]
    for n in [0,1,1.5,2,3,4,5]:
        
        order="n="+str(n)
        xi=0.0001
        theta=1.0
        f1=0.0
        theta_sol=[]
        xi_sol=[]
        
        for i in range(N):  #integrais interligados
            
            f1+= -xi**2 * (theta)**(n)*dxi
            theta+= f1/(xi**2)*dxi
            xi+=dxi
            theta_sol.append(theta)
            xi_sol.append(xi)

            """if(theta_sol[i]<0 and theta_sol[i-1]>=0 and n!=1.5):
                print(n,xi)
            if(theta_sol[i]>0 and theta_sol[i-1]<=0 and n!=1.5):
                print(n,xi)"""
        
        if(n==1.5):
            L1_5=theta_sol
        if(n==3):
            L3=theta_sol
             
        plot(xi_sol,theta_sol, label=order)
        legend(loc='lower left', fontsize=12)
        ylabel("Θ(ξ)", fontsize=30)
        xlabel("ξ", fontsize=30)
        xlim(0,10)
        ylim(-2,1)
        title("Numerical Solutions to Lane-Emden Equations", fontsize=25)
        axhline(y=0,c="black")
        
    return L3,L1_5

#f(N)
L3,L1_5=f(N)
show()
```

```python
L3_2=[k**3 for k in L3]
L1_5_2=[l**1.5 for l in L1_5]
qsi=[0.001*n for n in range(len(L3_2))]
plot(qsi,L3_2, label='n=3', linewidth=3)
plot(qsi,L1_5_2, label='n=1.5', linewidth=3)
legend(loc='upper right', fontsize=20)
ylabel(r"$Θ(ξ)^n$", fontsize=30)
xlabel("ξ", fontsize=30)
#xlim(0,10)
#ylim(-2,1)
title("Numerical Solutions to Lane-Emden Equations", fontsize=25)
axhline(y=0,c="black", linestyle='dashed')
```
